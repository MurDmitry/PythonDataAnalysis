Импорт необходимых библиотек

```python
import re
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Считывание всех строчек из файла n_log2.txt в lines
with open('n_log2.txt', 'r') as f:
    lines = f.readlines()

all_data = []  # Информация из строк, которые удовлетворяют условию

for line in lines:
    # .*? любое количество симолов между двумя частями (до и после .*?)
    match = re.search(r'(\d+:\d+:\d+,\d+).*?A00000000002 <--->.*?KEEP.*?volume=(\d+)', line)
    if match:
        all_data.append({'time': match.group(1), 'volume': match.group(2)})
print(all_data)
```

```python
data_graph_1 = []
for d in all_data:
    match = re.search(r'15:(0[0-9]:\d+,\d+|10:00,00)', d['time'])
    if match:
        data_graph_1.append({'time': d['time'], 'volume': d['volume']})
print(data_graph_1)
```

ВАЖНО! 2 графика должны быть вместе! См 2 графика вместе ниже

Первый график (отдельно)

```python
df = pd.DataFrame(data_graph_1)
print(df)
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S,%f')     # Преобразование колонки времени в datetime 
df['volume'] = pd.to_numeric(df['volume'])      # volume - это число
print(df)

# Отображение на графике
plt.figure(figsize=(10, 5))    # Размер графика 
plt.plot(df['time'], df['volume'])      # Значения на осях х (время) и y (volume)
plt.title('Первый график')
plt.xlabel('Время')
plt.ylabel('volume')    
plt.grid()                      # Сетка
plt.show()
```

ВАЖНО! 2 графика должны быть вместе! См 2 графика вместе ниже

Второй график (отдельно)

```python
df2 = pd.DataFrame(all_data)
df2['time'] = pd.to_datetime(df2['time'], format='%H:%M:%S,%f')     # Преобразование колонки времени в datetime
df2['volume'] = pd.to_numeric(df2['volume'])

print("============= Данные до обработки =============\n", df2)
df2.set_index('time', inplace=True)
print("============= set_index (time) =============\n", df2)
df_resampled2 = df2.resample('10min').mean()
print("============= Среднее за каждые 10 минут =============\n", df_resampled2)
df_resampled2['group'] = range(1, len(df_resampled2) + 1)     # Создание новой колонки с номерами групп
print("============= Добавление групп =============\n", df_resampled2)

# Отображение на графике
plt.figure(figsize=(10, 5))
plt.plot(df_resampled2['group'], df_resampled2['volume'])
plt.title('Среднее значение по 10 минутным интервалам')
plt.xlabel('Номер группы')
plt.ylabel('Среднее значение')
plt.grid()
plt.show()
```

Два графика вместе

```python
# Создание фигуры и осей
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# Построение первого графика
ax1.plot(df['time'], df['volume'], label='A000000002', color='blue')
ax1.set_title('Первый график')
ax1.set_xlabel('Время')
ax1.set_ylabel('volume')
ax1.legend()
ax1.grid()

# Построение второго графика
ax2.plot(df_resampled2['group'], df_resampled2['volume'], label='A000000002', color='red')
ax2.set_title('Второй график')
ax2.set_xlabel('Время')
ax2.set_ylabel('volume')
ax2.legend()
ax2.grid()

# Отображение графика
plt.tight_layout()  # Автоматическая настройка подгонки
plt.show()
```
